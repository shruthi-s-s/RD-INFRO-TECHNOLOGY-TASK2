```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/netservicechurnuse_cleaned.csv")
sns.set(style="whitegrid")
df
```

```python
df['churn_label'] = df['churn'].map({0: 'Stayed', 1: 'Churned'})
sns.countplot(data=df, x='churn_label', hue='churn_label', palette='muted',legend=False)
plt.title('Churn Count')
plt.xlabel('Customer Status')
plt.ylabel('Number of Customers')
plt.show()
```

```python
df['churn'].value_counts().plot(
    kind='pie',
    labels=['Churned', 'Stayed'],
    autopct='%1.1f%%',
    colors=['salmon','lightblue']
)
plt.title("Churn Rate")
plt.ylabel('')
plt.show()
```

```python
df['tv_sub_label'] = df['is_tv_subscriber'].map({0: 'Yes', 1: 'No'})
sns.countplot(data=df, x='tv_sub_label', hue='churn_label', palette='inferno',legend=False)
plt.title('TV Subscription vs Churn')
plt.xlabel('TV Subscriber')
plt.ylabel('Count')
plt.show()
```

```python
df['movie_pack_label'] = df['is_movie_package_subscriber'].map({0: 'No', 1: 'Yes'})
sns.countplot(data=df, x='movie_pack_label', hue='churn', palette='viridis')
plt.title('Movie Package vs Churn')
plt.xlabel('Movie Package')
plt.ylabel('Count')
plt.legend(title='Churn', labels=['Stayed', 'Churned'])
plt.show()
```

```python
sns.histplot(data=df, x='subscription_age', bins=30, kde=True, color='#720026')
plt.title('Subscription Age Distribution')
plt.xlabel('Months')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(data=df, x='churn', y='subscription_age', palette='pastel', hue='churn' ,legend=False)
plt.title('Subscription Age vs Churn')
plt.xlabel('Churn')
plt.ylabel('Subscription Age')
plt.xticks([0, 1], ['Stayed', ' Churned'])
plt.show()
```

```python
sns.violinplot(data=df, x='churn', y='bill_avg', palette='colorblind', hue='churn',legend=False)
plt.title('Bill Amount vs Churn')
plt.xlabel('Churn')
plt.ylabel('Average Bill')
plt.xticks([0, 1], ['Stayed', ' Churned'])
plt.show()
```

```python
sns.boxplot(data=df, x='churn', y='download_avg', hue='churn', palette='husl',legend=False)
plt.title('Download Average vs Churn')
plt.xlabel('Churn ')
plt.ylabel('Download Average')
plt.xticks([0, 1], ['Stayed', ' Churned'])
plt.show()
```

```python
sns.violinplot(data=df, x='churn', y='upload_avg', hue='churn', palette='Set1',legend=False)
plt.title('Upload Average vs Churn')
plt.xlabel('Churn')
plt.ylabel('Upload Average')
plt.xticks([0, 1], ['Stayed', ' Churned'])
plt.show()
```
